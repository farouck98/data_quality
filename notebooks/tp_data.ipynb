## Analyse des anomalies d'encodage

```python
# Recherche de caractères suspects pouvant révéler
# un problème d'encodage (UTF-8 / Latin-1 / CP1252)

colonnes_textes = df_silver.select_dtypes(include="object").columns

caracteres_suspects = ["Ã", "Â", "�", "Ä", "Ö", "Û"]

resultats_encodage = []

for col in colonnes_textes:

    masque = df_silver[col].astype(str).str.contains(
        "|".join(caracteres_suspects),
        regex=True,
        na=False
    )

    nb = masque.sum()

    if nb > 0:
        resultats_encodage.append(
            [col, nb]
        )

encodage_df = pd.DataFrame(
    resultats_encodage,
    columns=[
        "colonne",
        "nb_valeurs_suspectes"
    ]
).sort_values(
    "nb_valeurs_suspectes",
    ascending=False
)

encodage_df
```

### Normalisation des valeurs de condition_acces

```python
# Normalisation des variantes d'encodage

mapping_condition_acces = {
    "Accčs libre": "Accès libre",
    "Acc�s libre": "Accès libre",
    "Acc¸s libre": "Accès libre",
    "AccÃ¨s libre": "Accès libre"
}

df_silver["condition_acces"] = (
    df_silver["condition_acces"]
    .replace(mapping_condition_acces)
)
```

```python
df_silver["condition_acces"].value_counts(dropna=False)
```

toute valeur de condition_acces qui contient “libre” devient Accès libre, et toute valeur qui contient “réservé” reste Accès réservé.

```python
# Normalisation de la colonne condition_acces
# Objectif : regrouper toutes les variantes corrompues de "Accès libre"

df_silver["condition_acces"] = np.where(
    df_silver["condition_acces"].astype(str).str.contains("libre", case=False, na=False),
    "Accès libre",
    df_silver["condition_acces"]
)

df_silver["condition_acces"] = np.where(
    df_silver["condition_acces"].astype(str).str.contains("réservé|reservé|reserve|réserve", case=False, regex=True, na=False),
    "Accès réservé",
    df_silver["condition_acces"]
)

# Vérification après correction
df_silver["condition_acces"].value_counts(dropna=False)
```

```python
# Imputation locale des codes postaux manquants
# Principe : si plusieurs lignes ont les mêmes coordonnées GPS,
# et qu'une d'elles possède un code postal, on réutilise ce code postal.

cp_avant = df_silver["consolidated_code_postal"].isna().sum()

mapping_gps_cp = (
    df_silver
    .dropna(subset=[
        "consolidated_latitude",
        "consolidated_longitude",
        "consolidated_code_postal"
    ])
    .drop_duplicates(subset=[
        "consolidated_latitude",
        "consolidated_longitude"
    ])
    .set_index([
        "consolidated_latitude",
        "consolidated_longitude"
    ])["consolidated_code_postal"]
)

masque_cp_manquant = df_silver["consolidated_code_postal"].isna()

df_silver.loc[
    masque_cp_manquant,
    "consolidated_code_postal"
] = df_silver.loc[
    masque_cp_manquant
].set_index([
    "consolidated_latitude",
    "consolidated_longitude"
]).index.map(mapping_gps_cp)

cp_apres = df_silver["consolidated_code_postal"].isna().sum()

print("Codes postaux manquants avant :", cp_avant)
print("Codes postaux manquants après :", cp_apres)
print("Codes postaux imputés :", cp_avant - cp_apres)
```

Imputation à partir des coordonnées GPS existantes. Lorsque plusieurs bornes partagent les mêmes coordonnées GPS, le code postal d'une borne connue peut être réutilisé pour compléter les autres.
