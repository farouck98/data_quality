```python
# Analyse des identifiants répétés

# Nombre total de lignes du dataset
nb_lignes = len(df)

# Nombre d'identifiants de points de charge distincts
# dropna=True : les valeurs manquantes ne sont pas comptées
nb_id_pdc_uniques = df["id_pdc_itinerance"].nunique(dropna=True)

# Nombre d'identifiants manquants
nb_id_pdc_manquants = df["id_pdc_itinerance"].isna().sum()

# Nombre de lignes concernées par un identifiant répété
# keep=False marque toutes les occurrences comme doublons potentiels
nb_lignes_id_repetes = (
    df.duplicated(
        subset=["id_pdc_itinerance"],
        keep=False
    )
).sum()

# Nombre de répétitions supplémentaires
# C'est ce calcul qui correspond aux 67 094 mentionnés dans le sujet
nb_doublons = (
    df["id_pdc_itinerance"]
    .duplicated()
    .sum()
)

# Création du tableau récapitulatif
analyse_grain = pd.DataFrame({
    "indicateur": [
        "Nombre total de lignes",
        "Nombre d'id_pdc_itinerance uniques",
        "Nombre d'id_pdc_itinerance manquants",
        "Nombre de lignes avec id_pdc_itinerance répété",
        "Nombre de doublons (répétitions supplémentaires)"
    ],
    "valeur": [
        nb_lignes,
        nb_id_pdc_uniques,
        nb_id_pdc_manquants,
        nb_lignes_id_repetes,
        nb_doublons
    ]
})

analyse_grain
```

## Journal de décisions - Grille d'audit qualité

### Problème

67 094 répétitions de l'identifiant `id_pdc_itinerance`.

### Options envisagées

- Supprimer toutes les répétitions avec `drop_duplicates()`
- Conserver uniquement la version la plus récente
- Analyser le grain avant toute suppression

### Choix retenu

Analyser le grain avant toute suppression.

### Justification métier

L'étude d'un identifiant répété a montré que les lignes ne sont pas identiques et peuvent correspondre à plusieurs versions ou livraisons d'un même point de charge.

Le sujet indique explicitement que le grain du fichier est potentiellement « point de charge × livraison ». Une suppression automatique risquerait donc de supprimer des données valides.
